```python
from abc import ABC, abstractmethod

class Person(ABC):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    @abstractmethod
    def get_details(self):
        pass

class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)
        self.__grade = grade

    def set_grade(self, grade):
        if 1 <= grade <= 12:
            self.__grade = grade
        else:
            print("Grade must be between 1 and 12.")

    def get_details(self):
        return f"Student: {self.name}, Age: {self.age}, Grade: {self.__grade}"

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    
    def get_details(self):
        return f"Teacher: {self.name}, Age: {self.age}, Subject: {self.subject}"

def school_directory(people):
    for person in people:
        print(person.get_details())

# Testarea funcționalității
student1 = Student("Alice", 16, 10)
student2 = Student("Bob", 15, 9)

teacher1 = Teacher("Mr. Smith", 35, "Math")
teacher2 = Teacher("Ms. Johnson", 40, "Science")

people_list = [student1, student2, teacher1, teacher2]

school_directory(people_list)
```

```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Set the dimensions of the window
WIDTH, HEIGHT = 800, 600
win = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Golden Dash")

# Define colors
WHITE = (255, 255, 255)
YELLOW = (255, 255, 0)
RED = (255, 0, 0)

# Define player size
PLAYER_SIZE = 50

# Player class
class Player:
    def __init__(self, x, y, size, color):
        self.x = x
        self.y = y
        self.size = size
        self.color = color
        self.vel = 5

    def draw(self, win):
        pygame.draw.rect(win, self.color, (self.x, self.y, self.size, self.size))

# Function to create yellow squares
def create_yellow_square():
    x = random.randint(0, WIDTH - PLAYER_SIZE)
    y = 0
    return [x, y, YELLOW]

# Function to create red squares
def create_red_square():
    x = random.randint(0, WIDTH - PLAYER_SIZE)
    y = 0
    return [x, y, RED]

# Function to display score
def show_score(score):
    font = pygame.font.SysFont(None, 30)
    text = font.render(f"Score: {score}", True, WHITE)
    win.blit(text, (10, 10))

# Function to detect collisions and update score
def collision_detection(player, squares, score):
    for square in squares[:]:
        if square[1] + PLAYER_SIZE > player.y and square[0] + PLAYER_SIZE > player.x and square[0] < player.x + PLAYER_SIZE:
            squares.remove(square)
            if square[2] == YELLOW:
                score += 1
            elif square[2] == RED:
                return True, score
    return False, score

# Main game loop
def main():
    run = True
    clock = pygame.time.Clock()
    player = Player(WIDTH // 2, HEIGHT - 50, 50, (255, 255, 255))
    yellow_squares = []
    red_squares = []
    score = 0

    while run:
        clock.tick(60)
        win.fill((0, 0, 0))

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False

        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and player.x - player.vel > 0:
            player.x -= player.vel
        if keys[pygame.K_RIGHT] and player.x + player.vel < WIDTH - player.size:
            player.x += player.vel

        player.draw(win)

        # Generate yellow squares
        if random.randint(1, 100) == 1:
            yellow_squares.append(create_yellow_square())

        # Generate red squares
        if random.randint(1, 200) == 1:
            red_squares.append(create_red_square())

        # Move and draw yellow squares
        for square in yellow_squares[:]:
            square[1] += 1
            pygame.draw.rect(win, YELLOW, (square[0], square[1], PLAYER_SIZE, PLAYER_SIZE))
            if square[1] > HEIGHT:
                yellow_squares.remove(square)

        # Move and draw red squares
        for square in red_squares[:]:
            square[1] += random.randint(1, 3)
            pygame.draw.rect(win, RED, (square[0], square[1], PLAYER_SIZE, PLAYER_SIZE))
            if square[1] > HEIGHT:
                red_squares.remove(square)

        # Collision detection
        game_over, score = collision_detection(player, yellow_squares + red_squares, score)
        if game_over:
            # Game over screen
            while True:
                for event in pygame.event.get():
                    if event.type == pygame.QUIT:
                        pygame.quit()
                        quit()

                font = pygame.font.SysFont(None, 50)
                text = font.render("Game Over!", True, WHITE)
                win.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2 - text.get_height() // 2))

                font = pygame.font.SysFont(None, 30)
                text = font.render("Press R to restart or Q to quit", True, WHITE)
                win.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2 + text.get_height() // 2 + 10))

                keys = pygame.key.get_pressed()
                if keys[pygame.K_r]:
                    main()
                elif keys[pygame.K_q]:
                    pygame.quit()
                    quit()

                pygame.display.update()

        # Display score
        show_score(score)

        pygame.display.update()

    pygame.quit()

# Run the main function to start the game
if __name__ == "__main__":
    main()
```
